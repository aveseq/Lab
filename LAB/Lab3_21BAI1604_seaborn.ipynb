```python
import seaborn as sb
```

```python
print(sb.get_dataset_names())
```

```python
df1=sb.load_dataset("planets")
df1.info()
df1.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1['year'],df1['mass'])
plt.show()
```

```python
sb.relplot(x="year",y="distance",data=df1,kind='line',hue="number")
```

```python
sb.relplot(x="year",y="distance",data=df1,kind='line',col="number")
```

```python
import numpy as np
df2=np.random.randint(4,size=(2,5))
print(df2)
sb.heatmap(df2,annot=True)
```
