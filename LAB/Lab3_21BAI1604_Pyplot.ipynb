# Scatter Plot

```python

#Create lists 
x = np.arange(0, 200, 5)
y = np.arange(100, -100, -5)
y1 = np.arange(200, 0, -5)
plt.title("With labels and different colors")
plt.xlabel("Range 0 to 200")
plt.ylabel("Range 100 to -100")
plt.scatter(x, y, label="one", color='orange')
plt.scatter(x, y1, label="two", color='red')
plt.legend(['A', 'B'])
plt.savefig(r"plot3.jpg")
plt.show()
```

```python
#Create lists
x = np.arange(0, 200, 5)
y = np.arange(100, -100, -5)
y1 = np.arange(200, 0, -5)

plt.figure(figsize=(5, 5))
plt.title("With Gridlines and sized")
plt.xlabel("Range 0 to 200")
plt.ylabel("Range 100 to -100")

plt.grid(True)
plt.scatter(x, y, label="one", color='green')
plt.scatter(x, y1, label="two", color='cyan')
plt.legend(['A', 'B'])
plt.show()
```

# Twin axes plot


```python
figure=plt.figure()
axes1=figure.add_subplot(1,1,1)
axes2=axes1.twinx()

List1=np.linspace(0.,40.,400)
List2=[x**3 for x in List1]
List3=[x**4 for x in List1]

axes1.plot(List1,List2,'b-')
axes2.plot(List1,List3,'r-')

axes1.set_ylabel('x power 3',color="red")
axes2.set_ylabel('x power 5',color="blue")
axes1.set_xlabel('x')
```

# AI Vs ML Vs DL

```python
labels = ['AI Vs ML Vs DL']
sizes = [100]
fig, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

outer_circle = plt.Circle((0,0), 0.70, fc='green')
fig.gca().add_artist(outer_circle)

middle_circle = plt.Circle((0,0), 0.45, fc='orange')
fig.gca().add_artist(middle_circle)

inner_circle = plt.Circle((0,0), 0.20, fc='red')
fig.gca().add_artist(inner_circle)

ax1.axis('equal')  
plt.tight_layout()
plt.show()



```

# 3D Plotting

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

x=np.random.randint(10,size=(10))
y=np.random.randint(20,size=((10)))
z=np.random.randint(30,size=(10))

ax.scatter(x,y,z,c='purple',marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

# Subplots

```python
List1 = np.linspace(0., 10., 10)
List2 = [np.sin(x) for x in List1]
List3 = [x**3 for x in List1]
List4 = [np.tan(x) for x in List1]

fig = plt.figure(figsize=(20, 8))
plt.title("Welcome to subplot")
ax1 = fig.add_subplot(131)
ax3 = fig.add_subplot(133)
ax4 = fig.add_subplot(231)
ax5 = fig.add_subplot(232)
ax6 = fig.add_subplot(233)
ax1.plot(np.arange(10), np.arange(0, 100, 10), color='green')
ax3.plot(List1, List2, color='green')
ax5.plot(List1, List3, color='green')
ax6.plot(List1, List4, color='green')
ax4.bar(List1, List3, color='green')

plt.tight_layout()
plt.show()
```
