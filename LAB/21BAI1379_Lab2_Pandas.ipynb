```python
ipl_data = {'Team': ['KKR', 'KKR', 'RCB', 'RCB', 'CSK',
   'CSK', 'CSK', 'CSK', 'KKR', 'MI', 'MI', 'KKR'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
ipl = pd.DataFrame(ipl_data)

print(ipl)
```

```python
print(ipl.groupby('Team'))
```

```python
print(ipl.groupby('Team').groups)
```

```python
print(ipl.groupby(['Team','Year']).groups)
```

```python
grouped = ipl.groupby('Year')

for name,group in grouped:
  print(name)
  print(group)
```

```python
print(grouped.get_group(2014))
```

```python
grouped = ipl.groupby('Year')
print(grouped.get_group(2015))
```

```python
print(grouped['Points'].agg(np.mean))
print(grouped['Points'].agg(np.sum))
```
