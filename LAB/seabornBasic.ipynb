```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
print(sb.get_dataset_names())
```

```python
df=sb.load_dataset("flights")
df.info()
df.head()
```

```python
plt.scatter(df['year'],df['passengers'])
plt.show()
```

```python
sb.relplot(x="year",y="passengers",data=df,kind='line',hue="month")
```

```python
sb.relplot(x="year",y="passengers",data=df,kind='line',col="month")
```

```python
df1=np.random.randint(4,size=(2,5))
print(df1)
sb.heatmap(df1,annot=True)
```
