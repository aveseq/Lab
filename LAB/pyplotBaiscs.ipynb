```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line Plot
l1 = range(15)
l2 = [x*x for x in l1]
print(l1)
print(l2)
plt.plot(l1,l2)
plt.show()
```

```python
l1=np.arange(1,30,3)
l2=np.arange(1,30,3)
print(l1,"\n",l2,"\n")

plt.title("Line Plot")
plt.xlabel("Range 1-30 in steps of 3")
plt.ylabel("Range 1-30 in steps of 3")
plt.plot(l1,l2)
plt.show()
```

```python
l1=np.arange(1,20,0.2)
l2=np.sin(l1)
l3=np.cos(l1)
plt.title("Sin & Cos curve on List1(l1): ")
plt.xlabel("0 to 20 (0.2)")
plt.ylabel("-1 to 1")
plt.plot(l1,l2,l1,l3)
# plt.savefig(r"C:\users\)
plt.show()
```

**Scattter** **Plot**

```python
x = np.arange(0,100,2)
y = np.arange(50,-50,-2)
y1 = np.arange(100,0,-2)
plt.title("With labels & different colors") 
plt.xlabel("Range 0 to 50") 
plt.ylabel("Range 50 to -50 ") 
plt.scatter(x,y,label="one",color='green') 
plt.scatter(x,y1,label="one",color='orange') 
plt.legend(['A','B']) 
plt.show()
```

```python
x=np.arange(0,100,2)
y=np.arange(50,-50,-2)
y1=np.arange(100,0,-2)
plt.figure(figsize=(6,5))
plt.title("With Gridlines and sized")
plt.xlabel("Range 0 to 50")
plt.ylabel("Range 50 to -50 ")

plt.grid(True)
plt.scatter(x,y,label="one",color='green')
plt.scatter(x,y1,label="one",color='orange')
plt.legend(['A','B'])
#plt.savefig(r"C:\Users\plot3.jpg")
plt.show()
```

**TWIN AXES PLOT**

```python
figure=plt.figure()
axes1=figure.add_subplot(1,1,1)
axes2=axes1.twinx()

List1=np.linspace(0.,10.,100)
List2=[x**2 for x in List1]
List3=[x**3 for x in List1]

axes1.plot(List1,List2,'g-')
axes2.plot(List1,List3,'y-')

axes1.set_ylabel('x to power 2',color="yellow")
axes2.set_ylabel('x to power 3',color="green")
axes1.set_xlabel('x')
```

**Color, Marker & Line styles**

```python
from matplotlib import markers
x=np.linspace(-5.7,5.7,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,linewidth=5,c='black',linestyle='--')
plt.plot(x,z,c='green',marker='|',markersize=8)
plt.show()
```

**AI vs ML vs DL**

```python
labels=[' AI Vs ML Vs DL ']
sizes=[100]
fig,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
center_circle=plt.Circle((0,0),0.70,fc='pink')
fig=plt.gcf()
fig.gca().add_artist(center_circle)
inner_circle=plt.Circle((0,0),0.30,fc='red')
fig=plt.gcf()
fig.gca().add_artist(inner_circle)

ax1.axis('equal')
plt.tight_layout()
plt.show()
```

**3D Plotting**

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

x=np.random.randint(4,size=(10))
y=np.random.randint(16,size=((10)))
z=np.random.randint(12,size=(10))

ax.scatter(x,y,z,c='r',marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

**Subplots**

```python
Lt1=np.linspace(0.,10.,10)
Lt2=[x**2 for x in Lt1]
Lt3=[x**3 for x in Lt1]
Lt4=[np.sin(x) for x in Lt1]

fig=plt.figure(figsize=(20,8))
plt.title("Welcome to subplot")
ax1=fig.add_subplot(131)
#ax2=fig.add_subplot(132)
ax3=fig.add_subplot(133)
ax4=fig.add_subplot(231)
ax5=fig.add_subplot(232)
ax6=fig.add_subplot(233)
ax1.plot(np.arange(10),np.arange(0,20,2))
ax3.plot(Lt1,Lt2)
ax5.plot(Lt1,Lt3)
ax6.plot(Lt1,Lt4)
ax4.bar(Lt1,Lt2)
plt.tight_layout()
plt.show()
```
